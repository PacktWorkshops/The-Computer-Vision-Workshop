```python
# Load modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Load image
img = cv2.imread("../data/lion.jpg")
```

```python
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
# Get image shape
img.shape
```

```python
# Convert image to HSV
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(imgHSV[:,:,::-1])
plt.show()
```

```python
# Obtain the Hue channel
hue, sat, val = cv2.split(imgHSV)
```

```python
plt.imshow(val, cmap="gray")
plt.show()
```

```python
val = 255-val
```

```python
plt.imshow(val, cmap="gray")
plt.show()
```

```python
# Merge the val channels
imgHSV = cv2.merge((val,val,val))
```

```python
plt.imshow(imgHSV[:,:,::-1])
plt.show()
```

```python
# Flip the image horizontally
imgHSV = imgHSV[:,::-1,:]
```

```python
plt.imshow(imgHSV[:,:,::-1])
plt.show()
```

```python
# Create a new image
imgNew = np.zeros((407,640*2,3),dtype=np.uint8)
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
# Copy img to left half of new image
imgNew[:,:640,:] = img
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
# Copy imgHSV to right half of new image
imgNew[:,640:,:] = imgHSV
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
cv2.imwrite("mirror.png",imgNew)
```
