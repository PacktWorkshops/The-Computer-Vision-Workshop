```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the image
img = cv2.imread("../data/lion.jpg")
```

```python
# Display the image
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
# Find the shape of the image
img.shape
```

```python
# Create a new array with double the size
# Height will become twice
# Width and number of channels will
# stay the same
imgNew = np.zeros((814,640,3),dtype=np.uint8)
```

```python
# Display the image
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
# Copy the original image to the
# top half of the new image
imgNew[:407][:] = img
```

```python
# Display the image
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
# Invert the image
imgInverted = img[::-1,:,:]
```

```python
# Display the image
plt.imshow(imgInverted[:,:,::-1])
plt.show()
```

```python
# Copy the inverted image to the
# bottom half of the new image
imgNew[407:][:] = imgInverted
```

```python
# Display the image
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
# Save the image
cv2.imwrite("WaterEffect.png",imgNew)
```
